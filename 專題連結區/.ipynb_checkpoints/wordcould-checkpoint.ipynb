```python
import jieba
import jieba.analyse
import pandas as pd
import numpy
from collections import Counter
```

```python
from PIL import Image
import matplotlib.pyplot as plt
from wordcloud import WordCloud
mask = numpy.array(Image.open("1.png"))
font = "C:\Windows\Fonts\MSJH.TTC"
```

```python
orgdata = pd.read_csv('video_data.csv',encoding="utf-8")
```

```python
orgdata.head()
```

```python
orgdata['seg'] = ""
orgdata.head()
```

```python
jieba.analyse.set_stop_words('stopwords.txt')
index = 0
for documents in orgdata['Transcript'].astype(str):
    seg_list = jieba.analyse.extract_tags(documents, 20)
    #seg_list = jieba.lcut(documents)
    orgdata['seg'].iloc[index] = seg_list
    index += 1
```

```python
orgdata['seg']
```

```python
search_sub = orgdata[orgdata['seg'].apply(lambda x: '圖表' in x)]
search_sub
```

```python
show_sub = orgdata['Title']=='Excel 基礎教學 09：圖表製作 (上)'
subdata = orgdata[show_sub]
subdata_seg_list = subdata['seg'].tolist()
flat = [x for sublist in subdata_seg_list for x in sublist]
diction = Counter(flat)
wordcloud = WordCloud(background_color = "white", font_path=font, mask = mask)
wordcloud.generate_from_frequencies(diction)
plt.figure
plt.imshow(wordcloud)
plt.axis("off")
plt.savefig('L09.png', bbox_inches='tight')
plt.show()
```
