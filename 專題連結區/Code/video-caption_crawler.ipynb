```python
import yt_dlp

def download_subtitles(video_url):
    ydl_opts = {
        'writesubtitles': True,
        'allsubtitles': True,
        'skip_download': True,
        'subtitleslangs': ['zh-TW'],  # Optional: Specify the desired subtitle language (e.g., 'en' for English)
    }

    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        try:
            info_dict = ydl.extract_info(video_url, download=False)
            video_title = info_dict.get('title', None)
            subtitles = info_dict.get('subtitles', None)
            
            if subtitles:
                for lang, subtitle_list in subtitles.items():
                    for subtitle in subtitle_list:
                        subtitle_url = subtitle['url']
                        subtitle_filename = f'{video_title}_{lang}_{subtitle["ext"]}.srt'
                        ydl.download([subtitle_url])
                        print(f'Subtitle downloaded: {subtitle_filename}')
            else:
                print('No subtitles found for this video.')
                
        except yt_dlp.DownloadError as e:
            print('Download error:', str(e))
        except KeyError as e:
            print('Key error:', str(e))

# 输入 YouTube 视频链接
video_url = input('YouTube url: ')
download_subtitles(video_url)
```

```python
import jieba
import jieba.analyse
import numpy
from collections import Counter
from PIL import Image
import matplotlib.pyplot as plt
from wordcloud import WordCloud
mask = numpy.array(Image.open("1.png"))
font = "C:\Windows\Fonts\MSJH.TTC"
```

```python
seg_list = jieba.analyse.extract_tags(cc, 30)
seg_list
```

```python
diction = Counter(seg_list)

wordcloud = WordCloud(background_color="white", font_path=font, mask=mask)
wordcloud.generate_from_frequencies(diction)

plt.figure()
plt.imshow(wordcloud)
plt.axis("off")
plt.savefig('wordcloud.png', bbox_inches='tight')
plt.show()
```
