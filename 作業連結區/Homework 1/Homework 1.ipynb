# 各縣市社區大學資料比較

```python
import pandas as pd
from collections import Counter
import numpy as np
```

## Q1：110年度哪個縣市修課人次最高?

```python
df110A = pd.read_csv('110_Attendance.csv',encoding='big5')

df110A.head()
```

```python
Attendance_110= df110A[['縣市別', '修課人次合計']]
Attendance_110
```

```python
Attendance_110['修課人次合計'] = pd.to_numeric(Attendance_110['修課人次合計'].str.replace(',', ''))
```

```python
Attendance_110 = Attendance_110.groupby(['縣市別'])['修課人次合計'].sum().reset_index()
```

```python
Attendance_110['year'] = 110
Attendance_110
```

```python
Attendance_110.sort_values(by='修課人次合計',ascending=False).reset_index(drop=True)
```

## Q2：110年度哪個縣市社區大學據點最多?

```python
df110L = pd.read_csv('110_Location.csv',encoding='big5')
df110L.head()
```

```python
Location_110 = df110L.groupby(["縣市別"]).size().reset_index(name="據點數量")
Location_110['year'] = 110
Location_110
```

```python
Location_110.sort_values(by='據點數量',ascending=False).reset_index(drop=True)
```

## Q4：110年度哪個縣市社區大學擁有最多經費補助?

```python
CourseList = pd.merge(CourseList_110, CourseList_109, on='縣市別', how="outer")
CourseList = pd.merge(CourseList, CourseList_108, on='縣市別', how="outer")
CourseList.fillna(0, inplace=True)
CourseList
```

```python
cities_110_109_108 = []
cities_109_108_110 = []
cities_108_109_110 = []
cities_108_110_109 = []
cities_109_110_108 = []
cities_110_108_109 = []


for city in CourseList['縣市別']:
    # Get the course counts for the three years
    course_counts = CourseList.loc[CourseList['縣市別'] == city, ['108開課數量', '109開課數量', '110開課數量']].values[0]
    
    # Check if the course numbers follow the pattern of 108 > 109 > 110
    if all(x > y for x, y in zip(course_counts, course_counts[1:])):
        cities_108_109_110.append(city)
        
    # Check if the course numbers follow the pattern of 108 > 110 > 109
    elif course_counts[0] > course_counts[2] > course_counts[1]:
        cities_108_110_109.append(city)
    
    # Check if the course numbers follow the pattern of 109 > 108 > 110
    elif course_counts[1] > course_counts[0] > course_counts[2]:
        cities_109_108_110.append(city)
        
    # Check if the course numbers follow the pattern of 109 > 110 > 108
    elif course_counts[1] > course_counts[2] > course_counts[0]:
        cities_109_110_108.append(city)
    
    # Check if the course numbers follow the pattern of 110 > 109 > 108
    elif course_counts[2] > course_counts[1] > course_counts[0]:
        cities_110_109_108.append(city)
        
    # Check if the course numbers follow the pattern of 110 > 108 > 109
    elif course_counts[2] > course_counts[0] > course_counts[1]:
        cities_110_108_109.append(city)
```

```python
print("開課數量 108年度 > 109年度 > 110年度的縣市有: " + ", ".join(cities_108_109_110))

print("開課數量 108年度 > 110年度 > 109年度的縣市有: " + ", ".join(cities_108_110_109))

print("開課數量 109年度 > 108年度 > 110年度的縣市有: " + ", ".join(cities_109_108_110))

print("開課數量 109年度 > 110年度 > 108年度的縣市有: " + ", ".join(cities_109_110_108))

print("開課數量 110年度 > 109年度 > 108年度的縣市有: " + ", ".join(cities_110_109_108))

print("開課數量 110年度 > 108年度 > 109年度的縣市有: " + ", ".join(cities_110_108_109))
```
