# 各縣市社區大學修課人數
## 

```python
import pandas as pd

df = pd.read_csv('110.csv',encoding='big5')

df.head()
```

```python
temp = df[['縣市別', '修課人次合計']]
temp
```

```python
temp['修課人次合計'] = pd.to_numeric(temp['修課人次合計'].str.replace(',', ''))
```

```python
Type = temp.groupby(['縣市別'])['修課人次合計'].sum().reset_index()
```

```python
Type['year'] = 110
Type
```
